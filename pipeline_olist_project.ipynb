```python
import pandas as pd
import pandas._libs.tslibs as pdt
import sqlalchemy as sqa
import datetime as dt

customers_data = pd.read_csv("data/olist_customers_dataset.csv")
orders_data = pd.read_csv("data/olist_orders_dataset.csv") #
order_items_data = pd.read_csv("data/olist_order_items_dataset.csv")#
order_payments_data = pd.read_csv("data/olist_order_payments_dataset.csv") 
products_data = pd.read_csv("data/olist_products_dataset.csv")
sellers_data = pd.read_csv("data/olist_sellers_dataset.csv")
```

```python
order_items_data.columns
```

```python
orders_data.order_purchase_timestamp = orders_data.order_purchase_timestamp.apply(pd.to_datetime)
orders_data.order_purchase_timestamp
```

```python
orders_data.order_delivered_carrier_date = orders_data.order_delivered_carrier_date.apply(pd.to_datetime)
orders_data.order_delivered_carrier_date
```

```python
orders_data.order_delivered_customer_date = orders_data.order_delivered_customer_date.apply(pd.to_datetime)
orders_data.order_delivered_customer_date
```

```python
orders_data.order_estimated_delivery_date = orders_data.order_estimated_delivery_date.apply(pd.to_datetime)
orders_data.order_estimated_delivery_date
```

```python
orders_data.order_estimated_delivery_date = orders_data.order_estimated_delivery_date.apply(lambda x: pdt.Timestamp(x).date())
orders_data.order_estimated_delivery_date
```

```python
orders_data.order_approved_at = orders_data.order_approved_at.apply(pd.to_datetime)
orders_data.order_approved_at
```

```python
order_items_data.shipping_limit_date = order_items_data.shipping_limit_date.apply(pd.to_datetime)
order_items_data.shipping_limit_date
```

```python
engine = sqa.create_engine('postgresql+psycopg2://postgres:vincent23@localhost/olist_project')
```

```python
datasets_to_load = {'raw_customers':customers_data, 'raw_orders':orders_data, 'raw_order_items':order_items_data, 'raw_order_payments':order_payments_data, 'raw_products':products_data, 'raw_sellers':sellers_data}

for name, df in datasets_to_load.items():
    with engine.connect() as connection:
        df.to_sql(name, con = connection, if_exists = 'replace', index = False, chunksize = 10000)
```
